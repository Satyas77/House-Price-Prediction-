# Check Correlation & Drop Columns with Minute or No Correlation

```python
# Check for corrlation of Continious columns.

final_con_cols = list(df.corr()['SalePrice'].sort_values(ascending=False).head(20).index)
final_con_cols
```

```python
final_df_con = df[final_con_cols]
final_df_con.shape
```

```python
# Check for corrlation of Categorical columns.

cat = []

for i in df.columns:
    if df[i].dtype == object:
        cat.append(i)
```

```python
from repprep import ANOVA

final_cat_cols = []
for i in cat:
    print("--------------------")
    print("SalePrice vs",i)
    q = ANOVA(df,i,"SalePrice")
    print(q)
    if(q < 0.05):
        final_cat_cols.append(i)
```

```python
# Dividing data into predictors & target feature

x = final_df.drop(labels = 'SalePrice', axis = 1)

y = final_df[['SalePrice']]
```

# Standardization of Continuous columns and OHE of Categorical columns

```python
cat = []
con = []
for i in x.columns:
    if(x[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
import pandas as pd
X1 = pd.DataFrame(ss.fit_transform(x[con]),columns=con)
X2 = pd.get_dummies(x[cat])
x = X1.join(X2)
x.shape
```

# Divide Data in Training & Testing sets

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=31)
```

```python
xtrain.shape, x.shape
```

# Create Backward Elemination OLS Model

```python
#ytrain=ytrain.values.reshape(-1,1)

from statsmodels.api import OLS,add_constant
xconst = add_constant(xtrain)
ols = OLS(ytrain,xconst)
model = ols.fit()
model.summary()
```

# Remove Unncessary Columns on the basis of P-Value

```python
total_cols_dropped = []
```

```python
try:
    ar_before = round(model.rsquared_adj,4)

    Q = pd.DataFrame(model.pvalues.sort_values())
    col_to_drop = Q.tail(1).index[0]

    x = x.drop(labels=[col_to_drop],axis=1)
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=31)
    xconst = add_constant(xtrain)
    ols = OLS(ytrain,xconst)
    model = ols.fit()
    ar_after = round(model.rsquared_adj,4)

    if(ar_before > ar_after):
        print("Column dropped: ",col_to_drop)
        print("Column dropped incorrectly")
        x = x.join(final_df[[col_to_drop]])
        xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=31)
    else:

        total_cols_dropped.append(col_to_drop)
        print("Column dropped: ",col_to_drop)
        print("Column dropped Correctly")
        print("Prev ADJRsq:",ar_before)
        print("New ADJRsq:",ar_after)

    print('All columns dropped correctly.')

except KeyError:
    print('All columns dropped Correctly')
```

# Model_1 - Linear Regression

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

model = lm.fit(xtrain,ytrain)
pred_tr = model.predict(xtrain)
pred_ts = model.predict(xtest)

from sklearn.metrics import mean_absolute_error
tr_err = round(mean_absolute_error(ytrain,pred_tr),3)
ts_err = round(mean_absolute_error(ytest,pred_ts),3)
print("Training Error:",tr_err)
print("Testing Error:",ts_err)
```

# Model_2 - Ridge Regression

```python
from sklearn.linear_model import Ridge
rr = Ridge(alpha=10)
model = rr.fit(xtrain,ytrain)
pred_tr = model.predict(xtrain)
pred_ts = model.predict(xtest)
from sklearn.metrics import mean_absolute_error
tr_err = round(mean_absolute_error(ytrain,pred_tr),3)
ts_err = round(mean_absolute_error(ytest,pred_ts),3)
print("Training Error:",tr_err)
print("Testing Error:",ts_err)
```

```python
# Gridsearch CV - To find alpha value. 

W = []
x_cv = 1
for i in range(0,1000):
    W.append(round(x_cv+0.01,2))
    x_cv = x_cv + 0.01

tg = {"alpha":W}

rr = Ridge()
from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(rr,tg,scoring="neg_mean_absolute_error")
cvmodel = cv.fit(xtrain,ytrain)
cvmodel.best_params_
```
